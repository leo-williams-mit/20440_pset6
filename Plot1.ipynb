```python
import pandas as pd
import numpy as np
import bokeh.io
import bokeh.plotting
import iqplot
from bokeh.io import curdoc
curdoc().theme = 'dark_minimal'
bokeh.io.output_notebook()
```

```python
##Load the UQN counts file and set the gene names to be the index
file_path = 'Aurora_RNASeq_UQN_Counts_T.csv'
df = pd.read_csv(file_path)

df = df.set_index('Unnamed: 0')
df
```

```python
#Load the list of unique coding genes (as opposed to ribosomal rna or trna or other types of rna)
coding_genes_df = pd.read_csv('coding_genes.csv')
cod_genes = list(np.unique(coding_genes_df['gene_name']))
```

```python
#Load the metadata for each sample
meta = pd.read_csv('Aurora_Metadata.csv')
meta = meta.set_index('Unnamed: 0')
meta.head()
```

```python
##Isolate just the coding RNA, transpose the dataframe for formatting and add the metadata
df = df.loc[df.index.isin(cod_genes)]
counts = df.T
counts['Type'] = list(meta['Type'])
counts['Location'] = list(meta['Location'])
counts.head()
```

```python
#Load the toptable secured from the R limma analysis, and get just the coding genes
toptable = pd.read_csv('Prim-Meta_TopTable-True.csv')
toptable = toptable.set_index('X', drop=True)
toptable = toptable.loc[toptable.index.isin(cod_genes)]
toptable.head()
```

```python
#Isolate what's actually expressed, determine significant and nonsignificant fold changed genes. And among significant genes, split between upgregulated in metastatic and primary tumors
toptable = toptable.loc[toptable['AveExpr'] > 10]
nonsig = toptable.loc[(toptable['adj.P.Val'] > 0.05) | (np.abs(toptable['logFC']) < 1)]
sig = toptable.loc[(toptable['adj.P.Val'] < 0.05) & (np.abs(toptable['logFC']) > 1)]

high_met = sig.loc[sig['logFC'] < 0]
high_pri = sig.loc[sig['logFC'] > 0]
```

```python
#Make the plot
p = bokeh.plotting.figure(
    width=1000,
    height=1000,
    title='Volcano Plot of Primary vs. Metastasis',
    x_axis_label = 'log2(FC)',
    y_axis_label='-log10( Adjusted P Value)',


)

p.circle(
    x = nonsig['logFC'],
    y=-np.log10(nonsig['adj.P.Val']),
    color = 'gray'
)

p.circle(
    x = high_pri['logFC'],
    y=-np.log10(high_pri['adj.P.Val']),
    color = 'blue',
    legend_label = 'Upregulated in Primary Tumors'
)

p.circle(
    x = high_met['logFC'],
    y=-np.log10(high_met['adj.P.Val']),
    color = 'red',
    legend_label = 'Upregulated in Metastatic Tumors'
)

p.line(
    x=[-1.5e5, 1.5e5],
    y=[-np.log10(0.05), -np.log10(0.05)],
    line_dash = 'dashed'
)
p.line(
    x=[-1,-1],
    y=[0,12],
    line_dash='dashed'
)

p.line(
    x=[1,1],
    y=[0,12],
    line_dash='dashed'
)


bokeh.io.show(p)
```
